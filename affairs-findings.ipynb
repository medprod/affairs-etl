#### Demographics Analysis

##### 1. What is the comparison between the total percentage of women that cheated vs the men?

```python
query_gender = """
SELECT g.gender,
SUM(CASE WHEN df.cheated_flag THEN 1 ELSE 0 END) AS cheated_count,
ROUND((SUM(CASE WHEN df.cheated_flag THEN 1 ELSE 0 END)::NUMERIC /(SELECT COUNT(*) FROM demographics_fact))* 100, 2) 
AS percent_of_all,
ROUND((SUM(CASE WHEN df.cheated_flag THEN 1 ELSE 0 END)::NUMERIC /(SELECT COUNT(*) FROM demographics_fact WHERE cheated_flag)) * 100, 2) 
AS percent_of_cheaters
FROM demographics_fact df
JOIN gender_dim g ON df.gender_dim_id = g.gender_dim_id
GROUP BY g.gender;
"""
df_gender = run_query(query_gender)

df_gender
```

```python
plt.figure(figsize=(5, 5))
plt.pie(
    df_gender["percent_of_cheaters"],
    labels=df_gender["gender"],
    autopct="%.2f%%",
    colors=sns.color_palette("pastel"),
    startangle=90
)
plt.title("Cheating % by Gender")
plt.axis("equal")
```

##### 2. What is the comparison between the average number of affairs of same-sex vs opposite-sex couples?

```python
rship_query = """
SELECT 
    rt.relationship_type,
    COUNT(*) AS number_of_couples,
    SUM(df.affair_count) AS total_affairs,
    ROUND(AVG(df.affair_count),2) AS avg_num_of_affairs
FROM demographics_fact df
JOIN rship_type_dim rt 
    ON df.rship_type_dim_id = rt.rship_type_dim_id
GROUP BY rt.relationship_type;
"""

df_rship = pd.read_sql(rship_query, conn)

df_rship
```

```python
plt.figure(figsize=(6,4))
sns.barplot(
    data=df_rship,
    x="relationship_type",
    y="avg_num_of_affairs",
    palette="pastel"
)
plt.title("Average Number of Affairs by Relationship Type")
plt.ylabel("Average Number of Affairs")
plt.xlabel("Relationship Type")

# Add labels above bars
for i, row in df_rship.iterrows():
    plt.text(
        i, row.avg_num_of_affairs,
        str(row.avg_num_of_affairs),
        ha='center', va='bottom'
    )
plt.tight_layout()
plt.show()
```

#### Work & Education Analysis

##### 1. Are certain occupations more likely to cheat?

```python
occ_query = """
SELECT o.occupation_2025 AS occupation,
ROUND(AVG(we.affair_count),2) AS avg_affairs,
COUNT(*) AS respondent_count
FROM work_edu_fact we
JOIN occupation_dim o ON  we.occupation_dim_id = o.occupation_dim_id
GROUP BY o.occupation_2025
ORDER BY avg_affairs DESC;
"""

df_occ = pd.read_sql(occ_query, conn)
print(df_occ)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=df_occ,
    y="occupation", 
    x="avg_affairs", 
    palette="pastel"
)
plt.title("Average Number of Affairs by Occupation")
plt.xlabel("Average Number of Affairs")
plt.ylabel("Occupation")

# Add labels on bars
for index, row in df_occ.iterrows():
    plt.text(
        row.avg_affairs, index, 
        str(row.avg_affairs), 
        va='center'
    )

plt.tight_layout()
plt.show()
```

##### 2. Are people of certain education-levels more likely to cheat?

```python
edu_query = """
SELECT e.education_level_2025 AS education_level,
ROUND(AVG(we.affair_count),2) AS avg_affairs,
COUNT(*) AS respondent_count
FROM work_edu_fact we
JOIN education_dim e ON  we.education_dim_id = e.education_dim_id
GROUP BY e.education_level_2025
ORDER BY avg_affairs DESC;
"""

df_edu = pd.read_sql(edu_query, conn)
df_edu
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(
    data=df_edu,
    y="education_level",
    x="avg_affairs",
    palette="pastel",
    legend=False
)

plt.title("Average Number of Affairs by Education Level")
plt.xlabel("Average Affairs")
plt.ylabel("Education Level")

for index, row in df_edu.iterrows():
    plt.text(
        row.avg_affairs,  
        index,
        f"{row.avg_affairs:.2f}",
        va="center"
    )

plt.show()
```

#### Religion Analysis

##### 1. Does higher religiousness correlate to fewer affairs?

```python
rel_query = """
SELECT rel.religiousness_desc,
SUM(CASE WHEN ri.cheated_flag THEN 1 ELSE 0 END) AS num_cheated,
COUNT(*) AS total_respondents,
 ROUND(100.0 * SUM(CASE WHEN ri.cheated_flag THEN 1 ELSE 0 END) / COUNT(*), 2) AS percent_cheated
FROM religion_impact_fact ri
JOIN religiousness_dim rel ON ri.religiousness_dim_id = rel.religiousness_dim_id
GROUP BY rel.religiousness_desc
ORDER BY percent_cheated DESC;
"""

df_rel = pd.read_sql(rel_query, conn)
df_rel
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(
    data=df_rel,
    x="religiousness_desc",
    y="percent_cheated",
    palette="pastel",
    legend=False
)

plt.title("Percent Cheated by Religiousness Level")
plt.ylabel("Percent Cheated (%)")
plt.xlabel("Religiousness Level")
plt.xticks(rotation=45)

for index, row in df_rel.iterrows():
    plt.text(
        index,
        row.percent_cheated,
        f"{row.percent_cheated:.2f}%",
        ha='center'
    )

plt.tight_layout()
plt.show()
```

##### 2. Is there a correlation between religion and marriage satisfaction?

```python
relrat_query = """
SELECT rel.religiousness_desc,
ROUND(AVG(rat.rating_2025),2) AS avg_marriage_rating,
ROUND(AVG(rimp.affair_count),2) AS avg_num_affairs,
COUNT(*) AS respondent_count
FROM religion_impact_fact rimp
JOIN religiousness_dim rel ON rimp.religiousness_dim_id = rel.religiousness_dim_id
JOIN rating_dim rat ON rimp.rating_dim_id = rat.rating_dim_id
GROUP BY rel.religiousness_desc
ORDER BY rel.religiousness_desc;
"""

df_relrat = pd.read_sql(relrat_query, conn)
df_relrat
```

```python
fig, ax1 = plt.subplots(figsize=(8,5))

# Bar plot for marriage satisfaction
sns.barplot(
    data=df_relrat,
    x="religiousness_desc",
    y="avg_marriage_rating",
    hue="religiousness_desc",
    palette="pastel",
    ax=ax1,
    legend=False
)

ax1.set_ylabel("Average Marriage Rating", color="blue")
ax1.set_xlabel("Religiousness Level")
ax1.tick_params(axis='y', labelcolor="blue")

# Line plot for average number of affairs
ax2 = ax1.twinx()
sns.lineplot(
    data=df_relrat,
    x="religiousness_desc",
    y="avg_num_affairs",
    marker="o",
    color="red",
    ax=ax2
)
ax2.set_ylabel("Average Number of Affairs", color="red")
ax2.tick_params(axis='y', labelcolor="red")

plt.title("Marriage Satisfaction & Affairs by Religiousness")
plt.tight_layout()
plt.show()
```

```python
# Sort dataframe by avg_marriage_rating descending
df_relrat_sorted = df_relrat.sort_values("avg_marriage_rating", ascending=False)

# Make religiousness_desc a categorical with this order
df_relrat_sorted["religiousness_desc"] = pd.Categorical(
    df_relrat_sorted["religiousness_desc"],
    categories=df_relrat_sorted["religiousness_desc"],
    ordered=True
)

fig, ax1 = plt.subplots(figsize=(10,5))

# Bar plot for marriage satisfaction
sns.barplot(
    data=df_relrat_sorted,
    x="religiousness_desc",
    y="avg_marriage_rating",
    hue="religiousness_desc",
    palette="pastel",
    ax=ax1,
    legend=False
)
ax1.set_ylabel("Average Marriage Rating", color="blue")
ax1.set_xlabel("Religiousness Level")
ax1.tick_params(axis='y', labelcolor="blue")

# Line plot for average number of affairs
ax2 = ax1.twinx()
sns.lineplot(
    data=df_relrat_sorted,
    x="religiousness_desc",
    y="avg_num_affairs",
    marker="o",
    color="red",
    ax=ax2
)
ax2.set_ylabel("Average Number of Affairs", color="red")
ax2.tick_params(axis='y', labelcolor="red")

plt.title("Marriage Satisfaction & Affairs by Religiousness (Highest to Lowest Rating)")
plt.tight_layout()
plt.show()
```

#### Marriage Duration Analysis

##### Does the number of years a couple is married correlate to the number of affairs/overall satisfaction?

```python
years_query = """
SELECT mdf.years_married, 
ROUND(AVG(mdf.affair_count),2) AS avg_affairs,
ROUND(AVG(rat.rating_2025),2) AS avg_rating
FROM marital_duration_fact mdf
JOIN rating_dim rat ON mdf.rating_dim_id = rat.rating_dim_id
GROUP BY mdf.years_married
ORDER BY mdf.years_married;
"""

df_years = pd.read_sql(years_query, conn)
df_years
```

```python
fig, ax1 = plt.subplots(figsize=(8,5))

# Line for marriage rating
sns.lineplot(
    data=df_years,
    x="years_married",
    y="avg_rating",
    marker="o",
    color="blue",
    ax=ax1
)
ax1.set_ylabel("Average Marriage Rating", color="blue")
ax1.set_xlabel("Years Married")
ax1.tick_params(axis='y', labelcolor="blue")

# Secondary axis for average affairs
ax2 = ax1.twinx()
sns.lineplot(
    data=df_years,
    x="years_married",
    y="avg_affairs",
    marker="s",
    color="red",
    ax=ax2
)
ax2.set_ylabel("Average Number of Affairs", color="red")
ax2.tick_params(axis='y', labelcolor="red")

plt.title("Marriage Satisfaction & Affairs vs Years Married")
plt.tight_layout()
plt.show()
```
